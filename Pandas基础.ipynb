# 数据读取与写入
1.读取
    read_xxx(xxx是文件后缀)
    read_csv

2.写入
    to_xxx
    to_csv

```python
import pandas as pd

# 读取数据
df = pd.read_csv('data/sales.csv', usecols=['订单号', '产品类别', '产品名称','销售数量', '单价'])

# 数据处理
df['销售额'] = df['销售数量'] * df['单价']

# 写入数据     --> to_csv      --- index=False 不写入索引列
df.to_csv('data/sales_01.csv', index=False)
```

```python
# 数据查看 --> head, tail, describe, info, shape, columns
df1 = pd.read_csv('data/sales.csv', usecols=['订单号', '产品类别', '产品名称', '销售数量', '单价'])

df1.head(10)     # 显示前10行数据
df1.tail(10)     # 显示最后10行数据
print(df1.describe())   # 显示数据统计信息
print(df1.info())       # 展示数据信息（列名， 非空🧮，数据类型等）
df1.shape        # 显示数据行数和列数
df1.columns      # 显示列名
```

```python
# 数据选择
df2 = pd.read_csv('data/sales.csv', usecols=['订单号', '产品类别', '产品名称', '销售数量', '单价'], index_col="订单号")


# 1.选择列
# 1.1 单列
df2['产品名称']

# 1.2 多列
df2[['产品名称', '单价']]
df2[['产品类别', '产品名称', '单价']]

# 2. 选择行 - iloc, loc
# 2.1  iloc  -----> 基于行号选择行，语法: df.iloc[start:stop:step]
df2.iloc[0:5:1]      # 不包含结束位置
df2.iloc[:5]

# 2.2 loc    -----> 基于行索引选择行(包含结束位置)，语法: df.loc[start;stop;step]
df2.loc[6805677496:5897064217:1]
```
