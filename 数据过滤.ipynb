```python
import pandas as pd
df3 = pd.read_csv('data/sales.csv')
df3

# 配置项，展示所有数据
#pd.set_option('display.max_columns', None)

# 数据过滤 ---> df[filter]
# 1. 获取 销售数量 >= 10 的订单数据
df3[df3['销售数量'] >= 10]

# 2. 获取产品类别为 食品 或 图书 的订单数据
df3[df3['产品类别'].isin(['食品', '图书'])]

# 3. 获取 单价在100～200之间 的订单数据 ---> 范围：between ---> 包含边界
df3[df3['单价'].between(100, 200, inclusive = 'left')]
    # 包含左边,但不包含右边 （neither 都不取）

# 4. 获取 销售数量 >= 8, 并且 单价 >= 100 的订单数据 ---> 多条件(并且：&，或者：|)
df3[(df3['销售数量'] >= 8) & (df3['单价'] >= 100)]

# 5. 获取 产品类别为 服装/食品 ， 支付方式为 支付宝/微信支付 的订单数据
df3[(df3['产品类别'].isin(['服装', '食品'])) & (df3['支付方式'].isin(['支付宝', '微信支付']))]
```
