```python
import pandas as pd

pd.set_option('display.min_rows', 30)   # 设置显示行数(默认10行)

# 1. 读取数据
df = pd.read_csv('data/sales.csv')

# 2. 数据清洗
# 2.1 缺失值处理
df.isnull()   # 查看缺失值   True: 缺失

# 2.1.1  删除缺失值
df.dropna()     # 删除缺失值所在行
df.dropna(axis=1)   # 删除缺失值所在列
# /* dropna会返回DataFrame, 而不是在原DataFrame上作修改 */


# 2.1.2  填充缺失值
df.fillna('--') # 填充缺失值
df.ffill()  # 填充缺失值， 填充上一行数据
df.bfill()  # 填充缺失值， 填充下一行数据
```

```python
# 2.2 重复值处理
# 2.2.1 查看重复值
df.duplicated() # 查看重复值(所有的列的数据都重复)
df.duplicated(subset= ['订单号'])

# 2.2.2 删除重复值
df.drop_duplicates(subset=['订单号']) # keep='last' 保留重复值最后一行；keep='first' 保留重复值第一行
                                     # keep='false' 全删
```

```python
# 2.3 异常值处理
# 2.3.1 查看异常值
df[df['单价'] < 0]

# 2.3.2 删除异常值
print(df.drop(df[df['单价'] < 0].index))

# 2.3.3 修复异常值
df['单价'] = df['单价'].abs()   # 绝对值
```

```python
# 2.4 数据格式处理
df['订单日期'] = df['订单日期'].replace('2025/06/02', '2025-06-02')   # 只能替换单格数据，不推荐
df['订单日期'] = df['订单日期'].str.replace('/', '-')
```
