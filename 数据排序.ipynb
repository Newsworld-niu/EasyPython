```python
import pandas as pd

# 1. 读取数据
df = pd.read_csv('data/sales.csv', nrows=10)

# 2. 排序
# 2.1 根据 销售数量 倒序排序
df.sort_values('销售数量', ascending=False)   # 默认升序

# 2.2 根据 价格 升序排序
df.sort_values('单价', ascending=True)

# 2.3 根据 价格 升序排序， 价格一样， 再根据 销售数量 倒序排序
df.sort_values(['单价', '销售数量'], ascending=[True, False])
df.sort_values(['单价', '销售数量'], ascending=True)
```
